# Monitoreando ventas en el comercio

Bruno gestiona un pequeño comercio y quiere saber qué producto tuvo el mejor desempeño de ventas el mes pasado. Registró la cantidad vendida de dos productos: manzanas y plátanos. Ahora, necesita escribir un programa que identifique y muestre cuál de ellos tuvo más ventas.

Crea un programa que reciba el número de ventas de los dos productos y muestre un mensaje indicando cuál de ellos vendió más. Si las cantidades son iguales, muestra un mensaje diciendo que hubo un empate.

```python
manzanas = int(input("Cantidad de manzanas vendidas: "))
platanos = int(input("Cantidad de plátanos vendidos: "))

if manzanas > platanos:
  print("Manzanas tuvieron más ventas")
elif manzanas < platanos:
  print("Plátanos tuvieron más ventas")
else:
  print("Hubo misma cantidad de ventas")
```

# Calculando el tiempo total del proyecto
Camila está organizando un proyecto y necesita calcular el tiempo total necesario para concluir tres actividades: A, B y C. Sin embargo, si alguna actividad tiene un número de días negativo, el código debe avisar que los valores ingresados son inválidos y no calcular el total.

Escribe un programa que reciba el número de días de tres actividades y muestre el tiempo total del proyecto. Si algún valor es negativo, muestra un mensaje informando el error.

```python
actividad_a = int(input("Digite el numero de dias de la actividad A: "))
actividad_b = int(input("Digite el numero de dias de la actividad B: "))
actividad_c = int(input("Digite el numero de dias de la actividad C: "))

if actividad_a < 0 or actividad_b < 0 or actividad_c < 0:
  print("El tiempo estimado de las actividades no puede ser negativo")
else:
  tiempo_total = actividad_a + actividad_b + actividad_c
  print(f"El tiempo total del proyecto es de {tiempo_total} dias")
```

# Calculando el IMC
Anna Júlia está creando un sistema para calcular el Índice de Masa Corporal (IMC) y proporcionar recomendaciones básicas. El programa debe recibir el peso y la altura de una persona y mostrar el valor del IMC, además de indicar si está por debajo del peso, con peso normal o por encima del peso. Crea un programa que reciba el peso (en kg) y la altura (en metros) y calcule el IMC usando la fórmula: IMC = peso / (altura ** 2)Luego, muestra el valor del IMC y un mensaje indicando si está por debajo del peso (IMC < 18.5), peso normal (18.5 <= IMC < 25) o por encima del peso (IMC >= 25)

```python
peso = float(input("Digite su peso en kg: "))
altura = float(input("Digite su altura en metros: "))
IMC = peso / (altura ** 2)

if IMC < 18.5:
  print(f"Su IMC es {IMC}, usted esta por debajo del peso")
elif IMC <= 25:
  print(f"Su IMC es {IMC}, usted esta en el peso normal")
else:
  print(f"Su IMC es {IMC}, usted esta por encima del peso")
```

# Juego de múltiplos mágicos
Estás desarrollando un pequeño juego. El usuario ingresa un número entero y el programa debe evaluar lo siguiente:

Si el número es divisible por 3 y 5, muestra: "¡Número mágico!"

1.   Si solo es divisible por 3, muestra: "Divisible por 3"
2.   Si solo es divisible por 5, muestra: "Divisible por 5"
3.   Si no es divisible por ninguno, muestra: "No es un número mágico"

Este tipo de lógica es muy útil en juegos, validaciones o filtros.

```python
numero = int(input("Digite un número: "))

if numero % 3 == 0 and numero % 5 == 0:
  print(f"¡Número mágico!")
elif numero % 3 == 0:
  print(f"Divisible por 3")
elif numero % 5 == 0:
  print(f"Divisible por 5")
else:
  print(f"No es un número mágico")
```

# Sistema de clasificación para becas escolares
Una escuela otorga becas según tres criterios:



*   Ingreso familiar mensual.
*   Promedio del estudiante.
*   Asistencia (en porcentaje).


Reglas:


1.   Si el ingreso es menor a $1,500 y el promedio es mayor a 8.0 y la asistencia es al menos 90% → "Beca completa"

2.   Si el ingreso es menor a $2,500 y promedio mayor a 7.0 y asistencia al menos 85% → "Media beca"

3. En otros casos → "No elegible para beca"










```python
ingreso_mensual = float(input("Digite su ingreso mensual: "))
promedio = float(input("Promedio del estudiante: "))
asistencia = int(input("Asistencia del estudiante: "))

if ingreso_mensual < 1500 and promedio >= 8.0 and asistencia >= 90:
  print("Beca completa")
elif ingreso_mensual < 2500 and promedio >= 7.0 and asistencia >= 85:
  print("Beca parcial")
else:
  print("No elegible para beca")
```

# Sistema de tarifas de transporte por edad y distancia

Un sistema de transporte cobra según la edad del pasajero y la distancia recorrida:

* Menores de 6 años: Viajan gratis.
* De 6 a 18 años:
    Hasta 20 km: $1.50
    Más de 20 km: $2.50
* Mayores de 18:
    Hasta 20 km: $2.50
    Más de 20 km: $4.00

Crea un programa que reciba la edad y distancia, y muestre el valor a pagar.

```python
edad_pasajero = int(input("Digite su edad: "))
distancia = float(input("Digite la distancia en km: "))

if edad_pasajero < 6:
  print("Viaje gratis")
elif 6 <= edad_pasajero <= 18 and distancia <= 20:
  print("Valor del pasaje: $1.50")
elif 6 <= edad_pasajero <= 18 and distancia > 20:
  print("Valor del pasaje: $2.50")
elif edad_pasajero > 18 and distancia <= 20:
  print("Valor del pasaje: $2.50")
elif edad_pasajero > 18 and distancia > 20:
  print("Valor del pasaje: $4.00")
```

#Sistema de evaluación de desempeño empresarial


```python
ingresos_totales = int(input("Digite sus ingresos totales: "))
gastos_totales = int(input("Digite sus gastos totales: "))
numero_clientes_nuevos = int(input("Digite el numero de clientes nuevos: "))

if ingresos_totales - gastos_totales > 10000 and numero_clientes_nuevos > 50 :
  print("Trimestre Excelente")
elif ingresos_totales - gastos_totales > 5000 and numero_clientes_nuevos >= 20:
  print("Trimestre Bueno")
elif ingresos_totales - gastos_totales > 0 and numero_clientes_nuevos >= 10:
  print("Trimestre Regular")
else:
  print("Trimestre Deficitario")
```

## `IF` y `ELSE`

Los comandos `if` y `else` son dos estructuras condicionales. El comando `if` ejecutar√° el bloque de comandos en caso de que la condici√≥n citada sea  **verdadera**. El comando `else` ejecutar√° el bloque de comandos en el caso de que la condicional de `if` sea **falsa**.

El comando `if` es una palabra clave en Python que significa "si". Este se emplea para conformar la estructura condicional, que te permite verificar si una determinada condici√≥n es verdadera o falsa e, inmediatamente despu√©s, ejecute un bloque de c√≥digo espec√≠fico dependiendo del resultado de la verificaci√≥n. La sintaxis para usar el comando `if` es:



```
if condicion:
    # Realiza algo
```



````python
nota = float(input('Ingrese su nota: '))
if nota <= 10:
  print('Verdadero')
else :
  print('Falso')
````

Ya el comando `else` en Python se usa en junci√≥n con la palabra clave `if` para formar una estructura condicional. la sintaxis para usar el `else` es:

```
if condicion:
  # codigo en caso de que sea verdad
else:
  # codigo en caso de que sea falso
```

el comando `else` se ejecuta cuando la condici√≥n verificada por el `if` es evaluada como `False`.

## `ELIF`

El comando `elif` es una palabra clave en Python que significa "si no, si" y lo podemos considerar una *contracci√≥n* entre `else` e `if`. Se utiliza en conjunto con la palabra clave `if` para formar una estructura condicional en cadena.

la sintaxis para utilizar el comando `elif` es:

```
if condicion1:
    # Realiza algo
elif condicion2:
    # Realiza otra cosa
elif condicion3:
    # Realiza otra cosa
else:
    # Realiza algo diferente
```

El comando `elif` permite que puedas verificar varias condiciones en cadena, economizando espacio en tu c√≥digo. Si la primera condici√≥n se eval√∫a como `False`, el interpretador de Python evaluar√° la pr√≥xima condici√≥n con el `elif`. Ello continuar√° hasta que una condici√≥n sea evaluada como `True` o hasta que el `else` sea alcazado. Si ninguna de las condiciones es evaluada como `True`, la ejecuci√≥n del c√≥digo del comando `else` ser√° iniciada.

Vamos a emplear el mismo caso anterior:

````python
calificacion = float(input("Ingrese su nota: "))
if calificacion >= 7:
  print("aprobado")
elif calificacion >=5 and calificacion < 7:
  print("Recuperaci√≥n")
else:
  print("Reprobado")
````

## Operadores

Durante la construcci√≥n de comandos a veces necesitamos de una elaboraci√≥n mayor de la expresi√≥n condicional, necesitando que algunos operadores l√≥gicos se encuentren integrados.

### `AND`, `OR`, `NOT`

Los operadores l√≥gicos `and`, `or` y `not` son usados para combinar expresiones l√≥gicas en Python. Ellos son usados frecuentemente en conjunto con el comando `if` para crear estructuras condicionales m√°s complejas.


- `AND` es usado para verificar si dos condiciones son verdaderas. La expresi√≥n l√≥gica¬π `x and y` se eval√∫a como `True` tan solo si **ambas condiciones `x` y `y` son verdaderas**, y como `False` en caso contrario.

- `OR` es usado para verificar si al menos una de las condiciones es verdadera. La expresi√≥n l√≥gica `x or y` se eval√∫a como `True` **si al menos una de las condiciones `x` o `y` es verdadera**, y como `False` si ambas condiciones son falsas.

- `NOT` es usado para **negar una condici√≥n**. La expresi√≥n l√≥gica `not x` es evaluada como `True` si la condici√≥n x es falsa, y como `False` si la condici√≥n x es verdadera.

¬π Una expresi√≥n l√≥gica es una declaraci√≥n que puede ser evaluada como verdadera o falsa. Ella se compone por operandos l√≥gicos¬≤ y por operadores l√≥gicos¬≥, que son usados ‚Äã‚Äãpara combinar varias expresiones l√≥gicas en una √∫nica expresi√≥n.

¬≤ Los operandos l√≥gicos son los elementos que son comparados o evaluados en una expresi√≥n l√≥gica. Ellos son generalmente valores verdaderos o falsos, pero tambi√©n pueden ser expresiones l√≥gicas m√°s complejas. En Python, los operandos l√≥gicos son los valores `True` y `False`.

¬≥ Los operadores l√≥gicos son ls s√≠mbolos o palabras clave que son usados ‚Äã‚Äãpara combinar varias expresiones l√≥gicas en una √∫nica expresi√≥n. En Python, los operadores l√≥gicos son `and`, `or` y `not`, bien como las palabras clave `if`, `elif` e `else`.

# Estructuras de repetici√≥n

Cuando tenemos que ejecutar un mismo bloque de comandos durante varias ocasiones no es nada pr√°ctico hacerlo a mano.

**Situaci√≥n**: Nos fue solicitado obtener e imprimir el promedio de 2 notas de **3 estudantes**:

Ahora imaginemos una situaci√≥n en que no son tan solo 3 estudiantes, sino 100 estudiantes. No ser√≠a nada interesante repetir el mismo c√≥digo 100 veces, sino **ejecutar el mismo c√≥digo 100 veces**.

## `WHILE`

El lazo `while` es una estructura de control de repetici√≥n en Python que permite ejecutar un bloque de c√≥digo repetidamente mientras que una determinada condici√≥n sea verdadera. Su estructura es:



```
while condicion:
    # bloque de codigo
```

Vamos a construir un ejemplo con un contador de 1 hasta 10.


````python
contador = 1

while contador <= 10:
  print(contador)
  contador += 1
````

Ahora vamos a colectar las notas y promedios de cada alumno dentro del lazo `while`. Haremos un ejemplo con 3 promedios.

````python
contador = 1
while contador <= 3:
  nota_1 = float(input("ingrese la nota 1: "))
  nota_2 = float(input("ingrese la nota 2: "))
  print (f'El promedio del estudiante es: {(nota_1 + nota_2) / 2}')
  contador += 1
````

## `FOR`

El lazo `for` es un tipo de estructura de control de flujo en Python que permite iterar sobre un conjunto de elementos. Su estructura es:



```
for elemento in conjunto:
    # codigo a ser ejecutado para cada elemento
```

El lazo `for` itera sobre cada elemento del conjunto especificado y ejecuta el bloque de c√≥digo dentro del lazo para cada elemento. Cuando el lazo llega al final del conjunto, este se interrumpe y el programa contin√∫a la ejecuci√≥n despu√©s del lazo.



El conjunto lo podemos generar con la funci√≥n incorporada [`range()`](https://docs.python.org/3/library/functions.html#func-range). Se trata de una funci√≥n capaz de generar una secuencia de n√∫meros enteros. La estructura de esta funci√≥n es:

```
range(inicio, fin, paso)
```

Seg√∫n la documentaci√≥n, `range()` genera una secuencia de n√∫meros enteros a partir del valor del par√°metro `inicio` hasta el valor del par√°metro `fin`, de acuerdo con el valor del par√°metro `paso`. Si `inicio` no es especificado, el valor por defecto es `0`. Si `paso` no es especificado, el valor por defecto es `1`.




Vamos a recrear el mismo contador `while` con el lazo `for`.

````python
for e in range (3)
  nota_1 = float(input("ingrese la nota 1: "))
  nota_2 = float(input("ingrese la nota 2: "))
  print (f'El promedio del estudiante es: {(nota_1 + nota_2) / 2}')
````

````python
for i in range(1,11):
  if i == 4:
    continue
  print( i)
````

## Diccionario

Los diccionarios son un tipo de estructura de datos que almacenan pares de *llave-valor*. Estos son delimitados por llaves `{}` y los pares *llave-valor* son separados por comas.

```
diccionario = {llave: valor}
```

La **llave** es un elemento √∫nico que identifica a un valor en el diccionario, mientras que el **valor** es el item que ser√° almacenado para la llave. Las llaves y sus respectivos valores se pueden emplear para **cualquier tipo de dato**.

Los diccionarios son √∫tiles para almacenar y acceder a los datos de manera organizada y r√°pida. Se trata de un tipo de conjunto de elementos en Python, pues almacenan una colecci√≥n de items.

````python
diccionario = {
                "Perrito": "Aurelio",
                "Edad" : 9,
                "Raza" : "Chihuahua",
                "Sexo" : "Macho"
}
print(diccionario)

````

---
**Situaci√≥n:**

Vamos a crear un conjunto de datos con las informaciones de matr√≠cula de un estudiante. Los datos son los siguientes:

- matr√≠cula: 2000168933
- d√≠a de registro: 25
- mes de registro: 10
- grupo: 2E

````python
estudiante = {
    "matr√≠cula": 2000168933,
    "d√≠a de registro": 25,
    "mes de registro": 10,
    "grupo": "2E"
}
print(estudiante)
````

Es posible sustituir los valores dentro de una llave. Por ejemplo, recibimos la informaci√≥n de que el grupo del estudiante que registramos cambio a `'2G'` y ahora necesitamos cambiar el valor de la llave `'grupo'`.

````python
estudiante["grupo"] = "2G"
print(estudiante)
````

Tambi√©n podemos a√±adir otros datos al diccionario. Vamos a a√±adir la informaci√≥n sobre la modalidad de estudio, nuestro estudiante estudiar√° inicialemente en la modalidad EAD.

Definiremos una llave llamada `'modalidad'` con el valor `'EAD'`.

````python
estudiante["modalidad"] = "EAD"
print(estudiante)
````

### Lectura de valores con el lazo `for`

````python
tienda = {'nombres': ['televisi√≥n', 'celular', 'notebook', 'geladeira', 'estufa'],
          'precios': [2000, 1500, 3500, 4000, 1500]}
print(tienda)
````

````python
for clave, elementos in tienda.items():
  print(f'Clave: {clave}\nElementos:')
  for dato in elementos:
    print(dato)
````
